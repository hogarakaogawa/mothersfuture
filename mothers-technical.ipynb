```python
def BuyEntry(i, data): # è²·ã„ã‚·ã‚°ãƒŠãƒ«
    return data.iloc[i, 1] > data.iloc[i, 2] and data.iloc[i-1, 1] <= data.iloc[i-1, 2]

def SellEntry(i, data): # å£²ã‚Šã‚·ã‚°ãƒŠãƒ«
    return data.iloc[i, 1] < data.iloc[i, 2] and data.iloc[i-1, 1] >= data.iloc[i-1, 2]

def BuyExit(i, data): # è²·ã„ã‚¨ã‚¯ã‚¸ãƒƒãƒˆ
    return SellEntry(i, data)

def SellExit(i, data): # å£²ã‚Šã‚¨ã‚¯ã‚¸ãƒƒãƒˆ
    return BuyEntry(i, data)
```

```python
df['Long'] = 0
df['Sell'] = 0
lot = 1

for i in range(1, len(df.index)-1):
    if BuyEntry(i, df) and df.iloc[i, 4] == 0:
        df.iloc[i+1, 4] = lot
    elif BuyExit(i, df) and df.iloc[i, 4] != 0:
        df.iloc[i+1, 4] = 0
    else:
        df.iloc[i+1, 4] = df.iloc[i, 4]
        
    if SellEntry(i, df) and df.iloc[i, 5] == 0:
        df.iloc[i+1, 5] = lot
    elif SellExit(i, df) and df.iloc[i, 5] != 0:
        df.iloc[i+1, 5] = 0
    else:
        df.iloc[i+1, 5] = df.iloc[i, 5]
```

```python
df['BuyPoint'] = 0
df['SellPoint'] = 0
```

```python
df['BuyPoint'] = df['Long'].diff(1)
df['SellPoint'] = df['Sell'].diff(1)
```

```python
df
```

```python
LongPL = pd.Series(0.0, index=df.index)
SellPL = pd.Series(0.0, index=df.index)
spread = 1
Unit = 1000

BuyPrice = SellPrice = 0
BuyLot = SellLot = 0

BuyPoint = df['BuyPoint']
SellPoint = df['SellPoint']
```

```python
for i in range(1, len(df)):
    if BuyPoint[i] > 0:
        BuyPrice = df.iloc[i, 0] + spread
        BuyLot = df.iloc[i, 4] * Unit
        BuyPoint[i] = BuyPrice
    elif BuyPoint[i] < 0:
        ClosePrice = df.iloc[i, 0]
        LongPL[i] = (ClosePrice - BuyPrice) * BuyLot
        BuyPoint[i] = ClosePrice
    else:
        BuyPoint[i] = 'NaN'
        
    if SellPoint[i] > 0:
        SellPrice = df.iloc[i, 0]
        SellLot = df.iloc[i, 5] * Unit
        SellPoint[i] = SellPrice
    elif SellPoint[i] < 0:
        ClosePrice = df.iloc[i, 0] + spread
        SellPL[i] = (ClosePrice-SellPrice) * SellLot
        SellPoint[i] = ClosePrice
    else:
        SellPoint[i] = 'NaN'
```

```python
pd.DataFrame({'BuyPoint':BuyPoint, 'SellPoint':SellPoint, 'LongPL':LongPL, 'SellPL': SellPL})
```

```python
Equity = (LongPL + SellPL).cumsum()
```

```python
Equity.plot()
```
